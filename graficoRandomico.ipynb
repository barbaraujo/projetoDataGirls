```python
import random
import matplotlib.pyplot as plt

dados = [random.randint(1, 200) for _ in range(12)]

print("Lista de dados:", dados)

plt.plot(dados, marker ='o')
plt.title("\nGráfico de números aleatórios")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```
